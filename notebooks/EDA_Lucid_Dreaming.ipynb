```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='lucid_dream_freq', y='physical_activity')
plt.title("Physical Activity (mins/day) by Lucid Dream Frequency")
plt.xlabel("Lucid Dream Frequency")
plt.ylabel("Physical Activity (mins/day)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='lucid_dream_freq', y='sleep_hours')
plt.title("Sleep Hours by Lucid Dream Frequency")
plt.xlabel("Lucid Dream Frequency")
plt.ylabel("Sleep Hours")
plt.show()
```

### 🧠 Observations: Physical Activity & Sleep Hours vs Lucid Dreaming Frequency

Based on the visualizations of physical activity and sleep hours grouped by lucid dreaming frequency:

- **Activity Levels:**  
  Most lucid dreamers tend to get **60–90 minutes of physical activity per day**, but this range is also common among non-lucid dreamers.  
  This suggests that **physical activity alone may not be a reliable predictor** of lucid dreaming frequency.

- **Sleep Duration:**  
  - Lucid dreamers (monthly and weekly) typically sleep **7+ hours per night**, and their data spread is relatively **tight**, indicating more consistency.  
  - In contrast, the **'none' and 'nightly'** groups have **wider spreads**, especially the *'none'* group, which shows a lot of variability in sleep duration.  
  - This could suggest that **stable, consistent sleep patterns** may play a small role in supporting lucid dreaming.

- **Data Spread in Physical Activity:**  
  - The **widest spread** in physical activity is seen in **weekly dreamers**, followed by monthly dreamers.  
  - Nightly dreamers and non-dreamers show a more **compressed activity range**, though nightly dreamers lean toward **lower activity levels** overall.

> These patterns hint that a combination of **moderate activity and consistent sleep** may support lucid dreaming, but further analysis (e.g., multivariable interaction) would be needed to draw stronger conclusions.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

alcohol_ct = pd.crosstab(df['lucid_dream_freq'], df['alcohol_use'], normalize='index') * 100
marijuana_ct = pd.crosstab(df['lucid_dream_freq'], df['marijuana_use'], normalize='index') * 100

alcohol_plot = alcohol_ct.plot(kind='bar', stacked=True, ax=axes[0], colormap='Set2')
axes[0].set_title("Alcohol Use by Lucid Dream Frequency")
axes[0].set_xlabel("Lucid Dream Frequency")
axes[0].set_ylabel("Percentage (%)")
axes[0].legend(title="Alcohol Use")

for container in alcohol_plot.containers:
    alcohol_plot.bar_label(container, fmt='%.0f%%', label_type='center', fontsize=9)

marijuana_plot = marijuana_ct.plot(kind='bar', stacked=True, ax=axes[1], colormap='Set3')
axes[1].set_title("Marijuana Use by Lucid Dream Frequency")
axes[1].set_xlabel("Lucid Dream Frequency")
axes[1].legend(title="Marijuana Use")

for container in marijuana_plot.containers:
    marijuana_plot.bar_label(container, fmt='%.0f%%', label_type='center', fontsize=9)

plt.tight_layout()
plt.show()
```

### 🍷🌿 Substance Use Patterns by Lucid Dreaming Frequency

**Alcohol Use:**
- Lucid dreamers across all frequencies tend to fall in the **"rarely" to "weekly"** drinking range.
- The **"none"** group has a noticeably higher percentage of **daily drinkers**, possibly hinting at sleep disruption.

**Marijuana Use:**
- Nightly lucid dreamers are **more likely to report never or rarely using marijuana**, aligning with research suggesting marijuana can suppress REM sleep (where dreams occur).
- The **'weekly' lucid dreamers** have the widest spread in marijuana use patterns.

> Overall, the data suggests that **lower marijuana and alcohol use may be associated with more frequent lucid dreaming**, though more data would be needed for stronger conclusions.
